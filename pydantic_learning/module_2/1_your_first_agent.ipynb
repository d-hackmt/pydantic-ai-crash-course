# Your First Pydantic AI Agent

Now that you know how Pydantic makes sure data is strictly structured, let's see how Pydantic AI uses this to strictly control Large Language Models (LLMs).

We are going to use **Groq** to power our AI because it is extremely fast. 

> Note: In Jupyter environments running async code, we must patch the event loop with `nest_asyncio`.

```python
import nest_asyncio
nest_asyncio.apply()

import os
from dotenv import load_dotenv
from pydantic_ai import Agent

# Load environment variables (Make sure GROQ_API_KEY is set in your root .env file!)
load_dotenv()
```

### Agent 1: The Basic Agent

Let's create the simplest agent possible. It just takes a model name and nothing else.

```python
basic_agent = Agent('groq:llama-3.3-70b-versatile')

response = basic_agent.run_sync("Why is the sky blue? Answer in one short sentence.")

# print(response)

print("Basic Agent Answer:", response.output)
```

### Agent 2: The Persona Agent

A generic agent is boring. By giving an agent a **System Prompt**, we can completely change its personality and rules. Let's make an agent that only speaks like a Pirate.

```python
pirate_agent = Agent(
    'groq:llama-3.3-70b-versatile',
    system_prompt=(
        "You are Captain Monkey D Luffy. Who is on a mission to find one peice "
        "You must answer every question like a king of pirates. "
        "Never drop your character."
    )
)

response = pirate_agent.run_sync("Explain how to get food on a ship. Answer in one short sentence.")

print("Pirate Agent Answer:", response.output)
```
