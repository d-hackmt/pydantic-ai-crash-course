# Agent Observability with Logfire

LLMs can be incredibly opaque. When something goes wrong, it's hard to know if the prompt was bad, if the model hung, or if a validation schema failed. 

Because Pydantic AI and Logfire are built by the same core team, integration is entirely seamless. Let's see how.

```python
import nest_asyncio
nest_asyncio.apply()

from dotenv import load_dotenv

# Ensure your .env contains LOGFIRE_API_KEY
# Very Important: We must load the env BEFORE importing logfire!
load_dotenv()

import logfire
from pydantic_ai import Agent

# Configure Logfire to magically pick up the loaded token
logfire.configure()
logfire.instrument_pydantic_ai()
logfire.info('Logfire connected successfully! Hello {place}!', place='World')
```

### Running the Agent
We don't need to change our Agent code at all. Pydantic AI detects that Logfire is configured and automatically begins emitting telemetry data.

*(Warning: In Jupyter environments, if you aren't seeing traces immediately online, wait a moment or restart the kernel. Telemetry queues sometimes delay background flushing in notebooks. For 100% reliability, run the `.py` equivalent!)*

```python
basic_agent = Agent(
    'groq:llama-3.3-70b-versatile',
    system_prompt="You are a concise encyclopedia."
)

print("Sending request to LLM...")
response = basic_agent.run_sync("Give me 2 fast facts about black holes.")

print("\n--- Agent Response ---")
print(response.output)
```

### What just happened?

Go to your Logfire web dashboard, you will see a detailed trace of this execution. 
You can inspect exactly how many tokens were used, how long the Groq API took to respond, and the raw JSON schema that was sent over the wire!
